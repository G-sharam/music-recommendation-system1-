```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['text'])
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python

print(len(df[df['song'] == '']))
```

```python
def recommendation(song_name):
    df_reset = df.reset_index(drop=True) 
    matchingsong = df_reset[df_reset['song'] == song_name]

    if matchingsong.empty:
        print(" sorry your song is not detected.")
        return []

    idx = matchingsong.index[0]

    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])

    recommended_songs = []
    for i in distances[1:6]:
        recommended_songs.append(df_reset.iloc[i[0]]['song'])

    return recommended_songs
```

```python
recommendation('Oh, Pretty Woman')
```
